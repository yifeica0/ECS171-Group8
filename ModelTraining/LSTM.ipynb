```python


# DL models (CNNs, RNNs, LSTM)
from LSTM import LSTMClassifier
```

```python
# dataset
file_path = './datasets/amazon_user_reviews_text_sentiment_with_sw.parquet'
df = pd.read_parquet(file_path)
df.info()

# feature and target
features = "original_text"
target = "sentiment"
```

```python
# Model Training
# DL
print("DL: LSTM")

classifier = LSTMClassifier(vocab_size=10000, max_len=10, embedding_dim=128)
loss, accuracy = classifier.run_pipeline(
    df=df,
    text_column= features,
    label_column=target,
    epochs=20,
    batch_size=64,
    dropout_rate = 0.99
)
```
